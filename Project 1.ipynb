```python
#4. The age of most migrants

worldpop = pd.read_csv("world_population.csv")
worldpop['Med. Age'].mode()
```

```python
worldpop_plot = worldpop.sort_values(by='Population (2020)', ascending= False).head(10)
worldpop_plot
```

```python
sns.barplot(data=worldpop_plot, x='Density (P/Km²)', y='Med. Age')
plt.xticks(rotation= 45, horizontalalignment = 'right')# to prevent overlapping
plt.tight_layout()

plt.ticklabel_format(style='plain', axis='y')
```

```python
worldpop['Med. Age'].min()
```

```python
worldpop['Med. Age'].max()
```

```python
worldpop_plot=worldpop.sort_values(by='Fert. Rate', ascending= False).head(20)
worldpop_plot
```

```python
sns.barplot(data=worldpop_plot, x='Fert. Rate', y='Density (P/Km²)', errorbar=None)
plt.xticks(rotation= 45, horizontalalignment = 'right')# to prevent overlapping
plt.tight_layout()

plt.ticklabel_format(style='plain', axis='y')

# no relationship between high fertility rate and high population density
```

```python
worldpop_plot2=worldpop.sort_values(by='Fert. Rate', ascending= False).tail(20)
worldpop_plot2
```
